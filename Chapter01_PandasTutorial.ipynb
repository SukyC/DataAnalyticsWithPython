2. DataFrame의 병합

```python
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']},
                  index=[0,1,2,3])
df2=pd.DataFrame({'A':['A4','A5','A6','A7'],
                  'B':['B4','B5','B6','B7'],
                  'C':['C4','C5','C6','C7'],
                  'D':['D4','D5','D6','D7']},
                  index=[4,5,6,7])
df3=pd.DataFrame({'A':['A8','A9','A10','A11'],
                  'B':['B8','B9','B10','B11'],
                  'C':['C8','C9','C10','C11'],
                  'D':['D8','D9','D10','D11']},
                  index=[8,9,10,11])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
#데이터프레임을 열 방향으로 합치기(단순)
result = pd.concat([df1,df2,df3])
result
```

```python
#다중인덱스 형성
result = pd.concat([df1,df2,df3],keys=['x','y','z'])
result
```

```python
#생성된 인덱스를 확인해보자!
result.index
```

```python
result.index.get_level_values(0)
```

```python
result.index.get_level_values(1)
```

```python
df4=pd.DataFrame({'B':['B2','B3','B6','B7'],
                  'D':['D2','D3','D6','D7'],
                  'F':['F2','F3','F6','F7']},
                  index=[2,3,6,7])
```

```python
#세로가 아니라 가로로 붙이기
#concat은 인덱스를 기준으로 붙이기 때문에
#값을 넣을 수 없는 곳은 NaN
result=pd.concat([df1,df4],axis=1)
result
```

```python
#공통의 인덱스를 가진 값끼리만 합쳐주기
#join='inner' 옵션 사용!
result=pd.concat([df1,df4],axis=1,join='inner')
result
```

```python
#인덱스 무시하고 열방향으로 일단 합친 뒤
#새 인덱스 부여
result=pd.concat([df1,df4],ignore_index=True)
result
```

```python
#그냥 붙여도 컬럼명과 인덱스 참고해서 잘 붙긴 함.
result=pd.concat([df1,df4])
result
```

```python
left=pd.DataFrame({'key':['K0','K4','K2','K3'],
                  'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3']})
right=pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

```python
#공통된 key를 기준으로 합치기
pd.merge(left,right, on='key')
```

```python
#left 기준으로 합치기
pd.merge(left,right, how='left', on='key')
```

```python
#right 기준으로 합치기
pd.merge(left,right, how='right', on='key')
```

```python
#모두 합치기(합집합)
pd.merge(left,right, how='outer', on='key')
```

```python
#공통부분만 합치기(교집합)
pd.merge(left,right, how='inner', on='key')
```

# 01. CCTV와 인구 현황 파악하기

```python
cctv_seoul.head()
```

```python
#전체 cctv의 수가 가장 작은 5개의 자치구
cctv_seoul.sort_values(by='소계', ascending=True).head(5)
# 종로구, 도봉구, 중구, 노원구, 동작구
```

```python
#전체 cctv의 수가 가장 많은 5개의 자치구
cctv_seoul.sort_values(by='소계', ascending=False).head(5)
# 강남구, 관악구, 서초구, 은평구, 구로구
```

```python
#최근3년간 증가율 컬럼 생성하기
cctv_seoul['최근 증가율'] = cctv_seoul.loc[:,'2015년 이전':'2020년'].sum(axis=1) / cctv_seoul.loc[:,'2021년':'2023년'].sum(axis=1)
cctv_seoul.sort_values(by='최근 증가율', ascending=False).head(5)
# 영등포구, 강남구, 서대문구, 관악구, 성동구
```

```python
pop_seoul2.head()
```

```python
#문자열 타입이 된 숫자를 숫자형으로 변경
pop_seoul2[['인구수','한국인','외국인','고령자']]=pop_seoul2[['인구수','한국인','외국인','고령자']].apply(pd.to_numeric)
pop_seoul2.info()
```

```python
#구별 외국인 비율과 고령자비율
pop_seoul2['외국인비율']=pop_seoul2['외국인']/pop_seoul2['인구수'] * 100
pop_seoul2['고령자비율']=pop_seoul2['고령자']/pop_seoul2['인구수'] * 100
pop_seoul2.head()
```

```python
pop_seoul2.sort_values(by='외국인', ascending=False).head()
#외국인 많은 자치구 Top5: 구로구, 영등포구, 동대문구, 관악구, 광진구
```

```python
pop_seoul2.sort_values(by='외국인비율', ascending=False).head()
#외국인비율 높은 자치구 Top5: 중구, 종로구, 용산구, 영등포구, 금천구
```

```python
#한글폰트 사용 및 마이너스 부호 깨짐 방지
import matplotlib.pyplot as plt
from matplotlib import rc
plt.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
```

```python
data_result.head()
```

```python
data_result['소계'].plot(kind='barh', grid=True, figsize=(6,6))
plt.show()
```

```python
data_result['소계'].sort_values().plot(kind='barh', grid=True, figsize=(6,6))
plt.title('구별 CCTV 설치 대수')
plt.savefig('구별 CCTV 설치 대수.png')
plt.show()
#cctv 개수로는 강남구가 월등 그 뒤로 관악구, 서초구, 은평구, 구로구 많다.
```

```python
data_result['CCTV비율'] = data_result['소계']/data_result['인구수'] * 100
data_result['CCTV비율'].sort_values().plot(kind='barh', grid=True, figsize=(6,6))
plt.title('인구수 대비 구별 CCTV 설치 비율')
plt.savefig('인구수 대비 각 구별 CCTV 설치 비율.png')
plt.show()
#성동구가 제일 많다
#강서구, 송파구, 노원구는 많이 적다.
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV수')
plt.grid()
plt.show()
```

```python
#데이터를 대표하는 직선 그리기
fp1=np.polyfit(data_result['인구수'], data_result['소계'], 1)
fp1
```

```python
#y축 데이터
f1=np.poly1d(fp1)
#x축 데이터
fx=np.linspace(100000,700000, 100)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw = 3, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV수')
plt.grid()
plt.show()
```

```python
data_result['오차']=np.abs(data_result['소계'] - f1(data_result['인구수']))
```

```python
df_sort=data_result.sort_values(by='오차',ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구수'], data_result['소계'],c=data_result['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw = 3, color='g')

for n in range(10):
  #plt.text(x, y, s) text가 표시될 위치
  plt.text(df_sort['인구수'][n]*1.02, df_sort['소계'][n]*0.98,
           df_sort.index[n], fontsize=15)

plt.xlabel('인구수')
plt.ylabel('CCTV수')
plt.title('구별 인구수 대비 CCTV 설치 대수',fontsize=17)

plt.colorbar()
plt.grid()
plt.savefig('구별인구수대비cctv설치대수_final.png')
plt.show()
```

```python
import pandas as pd
import numpy as np
```

```python
#pandas의 기초 데이터 형태 Serise
s=pd.Series([1,3,np.nan,6,8])
s
```

```python
#pandas의 날짜형 데이터 형태
dates=pd.date_range('20130101', periods=6)
dates
```

```python
#데이터프레임 만들기, 인덱스 지정 및 컬럼명 지정
df=pd.DataFrame(np.random.randn(6,4), index=dates, columns=["A","B","C","D"])
df
```

```python
#기본 첫 5행 보여줌.
df.head()
```

```python
#몇개의 행을 볼것인지 지정 가능
df.head(3)
```

```python
#인덱스만 보기
df.index
```

```python
#컬럼명만 보기
df.columns
```

```python
#데이터프레임 안에 들어가는 내용물 보기
df.values
```

```python
#데이터프레임의 개요 확인하기
df.info()
```

```python
#데이터프레임의 통계적 개요 확인
df.describe()
```

```python
#지정 컬럼을 중심으로 행 정렬하기
#by 뒤에 컬럼명 넣음
df.sort_values(by='B', ascending = False)
```

```python
#컬럼만 시리즈 형태로 보기
df['A']
```

```python
#보고싶은 행의 범위 지정
df[0:3]
```

```python
#인덱스명으로 행을 지정할 수도 있음
df['2013-01-02':'2013-01-04']
```

```python
#위치를 기준으로 슬라이싱 하기
#loc() 의 활용, ()안에 인덱스 이름을 넣어준다
df.loc[dates[0]]
```

```python
#loc(행(row)의 범위, [열(col)의 범위])
df.loc[:,['A','B']]
```

```python
#인덱스 이름, 컬럼 이름으로 슬라이싱, 컬럼은 리스트 형태
df.loc['20130102':'20130104',['A','C']]
```

```python
df.loc['20130102',['A','C']]
```

```python
df.loc[dates[0],'A']
```

```python
#행, 열의 번호로 슬라이싱 하기
#iloc[]
df.iloc[3] #3번행만 선택됨(4번째 행)
```

```python
#iloc[행범위,열범위]
df.iloc[3:5,0:2]
```

```python
#불연속적인 범위는 리스트로 지정 가능
df.iloc[[1,2,4],[0,2]]
```

```python
#전체를 가져올 땐 콜론(:) 사용
df.iloc[1:3,:]
```

```python
df.iloc[:,1:3]
```

```python
#조건을 만족하는 행만 출력
df[df.A>0]
```

```python
#데이터프레임 전체에 조건을 걸면 불만족한 값은 NaN으로 출력됨
df[df>0]
```

```python
#데이터의 내용까지 복사하기
df2=df.copy()
df2
```

```python
#새로운 컬럼과 값 추가하기
df2['E'] = ['one','one','two','three','four','three']
df2
```

```python
#value가 있는지 확인하고 싶을때
df2['E'].isin(['two','four'])
```

```python
#위의 조건에 맞는 행만 출력
df2[df2['E'].isin(['two','four'])]
```

```python
#복사한 df2에 E를 추가했기 때문에 df는 그대로인 것을 확인 가능
df
```

```python
#apply()를 통해 데이터프레임에 함수를 적용할 수 있어
#np.cumsum 누적합 확인하기
df.apply(np.cumsum)
```

```python
#apply, lambda를 이용해서 할수도 있음
#최대값과 최소값의 차이 알기
df.apply(lambda x: x.max()-x.min())
```
